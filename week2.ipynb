```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = cv.imread('/content/drive/MyDrive/PCVK/Images/kitten01.png')

B, G, R = cv2.split(image)

RB = cv2.merge([B, np.zeros_like(G), R])

GB = cv2.merge([B, G, np.zeros_like(R)])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(RB, cv2.COLOR_BGR2RGB))
plt.title('Red-Blue Channel')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(GB, cv2.COLOR_BGR2RGB))
plt.title('Green-Blue Channel')

plt.show()
```

```python
cropped_image = image[20:115, 25:120]

plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image (Rows 20-115, Columns 25-120)')
plt.show()
```

```python
# Ekstrak channel merah dan tampilkan hanya baris 5-30 dari semua kolom
red = R[5:30, :]  # Baris 5-30, semua kolom

plt.imshow(red, cmap='Reds')
plt.title('Red Channel (baris 5-30)')
plt.show()
```

```python
# Ukuran gambar
height, width = 500, 500
image = np.ones((height, width, 3), dtype=np.uint8) * 255

# Gambar 5 kotak dengan ukuran dan posisi acak
for _ in range(5):
    start_point = (np.random.randint(0, width), np.random.randint(0, height))
    end_point = (start_point[0] + np.random.randint(10, 100), start_point[1] + np.random.randint(10, 100))
    color = (np.random.randint(0, 256), np.random.randint(0, 256), np.random.randint(0, 256))
    thickness = np.random.randint(1, 5)
    image = cv2.rectangle(image, start_point, end_point, color, thickness)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Random Boxes')
plt.show()
```

```python
# Membalikkan gambar
flipped_img_vertical = cv.flip(image, 1)
# Menampilkan gambar terbalik
plt.figure(figsize=(5, 5))
plt.imshow(cv.cvtColor(flipped_img_vertical, cv.COLOR_BGR2RGB))
plt.show()
```

```python

import matplotlib.patches as patches

# Membaca gambar
img = cv.imread('/content/drive/MyDrive/PCVK/Images/fotoku.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Menyiapkan figure dan axes dengan ukuran yang lebih besar
fig, ax = plt.subplots(figsize=(10, 5))

# Menampilkan gambar
ax.imshow(img_rgb)

# Koordinat dan ukuran rectangle
rect_start_point = (650, 330)
rect_width = 120
rect_height = 130
rect_color = 'green'
rect_linewidth = 2

# Menambahkan rectangle ke plot
rect = patches.Rectangle(rect_start_point, rect_width, rect_height, linewidth=rect_linewidth, edgecolor=rect_color, facecolor='none')
ax.add_patch(rect)

# Koordinat dan radius circle
circle_center = (710, 400)
circle_radius = 55
circle_color = 'red'
circle_linewidth = 2

# Menambahkan circle ke plot
circle = patches.Circle(circle_center, circle_radius, linewidth=circle_linewidth, edgecolor=circle_color, facecolor='none')
ax.add_patch(circle)

# Menampilkan gambar dengan overlay
plt.show()
```

```python
# Membaca citra
img = cv.imread('/content/drive/MyDrive/PCVK/Images/kitten01.png')

# Memisahkan channel B
b_channel = img[:, :, 0]

# Menentukan posisi dan ukuran rectangle
height, width = b_channel.shape
start_point = (0, height - 300)
end_point = (300, height)
color = 255
thickness = -2

# Menggambar rectangle pada channel B
b_channel_with_rectangle = b_channel.copy()
cv.rectangle(b_channel_with_rectangle, start_point, end_point, color, thickness)

# Menggabungkan kembali channel B ke dalam citra RGB
img_with_rectangle = img.copy()
img_with_rectangle[:, :, 0] = b_channel_with_rectangle

# Menambahkan teks pada gambar
text = 'KucingKu'
font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 4
font_color = (120, 12, 0)
thickness = 7
text_size, _ = cv.getTextSize(text, font, font_scale, thickness)
text_width, text_height = text_size

# Posisi teks di kanan bawah
position = (width - text_width - 10, height - 40)
cv.putText(img_with_rectangle, text, position, font, font_scale, font_color, thickness)

# Mengubah citra dari BGR ke RGB
img_with_rectangle_rgb = cv.cvtColor(img_with_rectangle, cv.COLOR_BGR2RGB)

# Menampilkan citra menggunakan matplotlib dengan ukuran lebih kecil
plt.figure(figsize=(6, 4))
plt.imshow(img_with_rectangle_rgb)
plt.show()
```
